```python
import mlxtend
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from apyori import apriori
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules 
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
           ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
           ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
           ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
           ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    
df=pd.DataFrame(te_ary, columns=te.columns_)  
df2 = df.copy()
df2.head()
```

```python
df2.info()
df2.shape
```

```python
apriori(df2, min_support=0.6)
```

```python
frequent_itemsets=apriori(df2, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)
```

```python
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

```python
le support n’est rien d’autre que le rapport entre le nombre d’opérations dans lesquelles l'element  x apparaît et le nombre total de transactions.

La confidence (x => y) signifie la probabilité que l’article y soit acheté lorsque l’article x est acheté. Cette méthode prend en compte la popularité de l’élément x.

L’ascenseur(lever) (x => y) n’est rien d’autre que « l’intérêt » ou la probabilité que l’article y soit acheté lorsque l’article x est vendu. Contrairement à la confiance (x => y), cette méthode tient compte de la popularité de l’élément y.
```
